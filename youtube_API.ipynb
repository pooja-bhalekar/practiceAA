```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By


def get_driver():
    """
    This will Initialise the webdriver and return Driver.
    :return:
    """
    # set Options to make browsing Easy!!
    chrome_driver = Options()
    chrome_driver.add_argument("--disable-infobars")
    chrome_driver.add_argument("--no-sandbox")
    chrome_driver.add_argument("--headless")
    chrome_driver.add_argument("--disable-dev-shm-usage")
    chrome_driver.add_argument("--disable-blink-features=AutomationControlled")
    web_driver = webdriver.Chrome("D:\\Python_ML_DL\\chromedriver.exe")
    return web_driver
```

```python
driver = get_driver()
```

```python
driver.get("https://www.accessdata.fda.gov/scripts/cder/daf/index.cfm")
```

```python
driver.find_element(By.ID, value="searchterm").send_keys("mitomycin")
```

```python
driver.find_element(By.XPATH, value='//*[@id="DrugNameform"]/div[2]/button[1]').click()
```

```python
pharma_details_perpage = driver.find_elements(By.CLASS_NAME, value="col-md-12")
```

```python
all_product_href = []
for i in range(len(pharma_details_perpage)):
    product_sublinks = pharma_details_perpage[i].find_element(By.TAG_NAME, "ul").find_elements(By.TAG_NAME,"li")#[0].find_element(By.TAG_NAME, "a").get_attribute("href")
    for j in range(len(product_sublinks)):
        all_product_href.append(product_sublinks[j].find_element(By.TAG_NAME, "a").get_attribute("href"))
```

```python
all_product_href
```

```python
url = all_product_href[0]
```

```python
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
page = requests.get(url=url, headers=headers)
print(page.status_code)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
len(soup.find_all('div', {"class":"panel panel-info"}))
```

```python
pdf_href = soup.find_all('div', {"class":"panel panel-info"})[1].find('div', {"class":"panel-body"}).find_all('a')
```

```python
hrefs_list = []
for i in range(len(pdf_href)):
    hrefs_list.append(pdf_href[i].get('href'))
```

```python
page_pdf = requests.get(hrefs_list[0], stream=True)
with open(f"{hrefs_list[0].split('/')[-1]}.pdf", 'wb') as fd:
    for chunk in page_pdf.iter_content(10,000):
        fd.write(chunk)
```
